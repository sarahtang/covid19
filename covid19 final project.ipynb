## Predict number of Covid-19 cases
In this section, I create linear models to produce estimates of the number of covid-19 cases on 4/18/20 by county. I conduct data cleaning in order to prepare the dataset for training and testing and also conduct feature engineering to get fastest growth rate.

I wanted to know what factors in a county are beneficial in predicting coronavirus cases and how fast the virus spreads in an area.

### Covid confirmed and county data and % change feature engineering

```python
# Combine covid confirmed and county data
county_covid = covid_confirmed.dropna(subset = ['Admin2'])
county_data_confirmed = abdg_counties.merge(county_covid, how = 'left',
                                            left_on = ['CountyName', 'State'],
                                            right_on = ['Admin2', 'Province_State'])
county_data_confirmed = county_data_confirmed.dropna(subset = ['4/18/20'])
county_data_confirmed
```

```python
# Get covid cases by date per county
county_data_confirmed_ind = county_data_confirmed.set_index(['CountyName', 'State'])
county_data_confirmed_dates = county_data_confirmed_ind[county_data_confirmed_ind.columns.intersection(list_dates)]
county_data_confirmed_dates.head()
```

```python
# No NAs in covid cases
county_data_confirmed_dates.isna().sum().sum()
```

Use covid-19 cases by day to determine percent change and find in which places is covid growing the most rapidly or slowly.

```python
# Transpose so that can find percent change
county_dates = county_data_confirmed_dates.transpose()
county_pct_change = county_dates.pct_change()
county_pct_change.tail()
```

Next, I decided to get the highest percentage change for each county to see if that would be an important feature in predicting the number of cases on 4/18/20. I think this will be important because as of 4/17/20, many counties have not yet experienced a decline in cases and are still in times of growth. Thus, the max percent change will be recent.

```python
# Get highest percent change for each county/state
# Getting infinity values where jumps from 0 to x number of cases 
# because x-0/0 = inf
county_pct_change.max(axis = 0)
```

```python
# so lots of one values
# getting values of up to 1, 2, 0, 3
county_pct_change.isin([np.inf]).sum().unique()
(county_pct_change.isin([np.inf]).sum() == 0).sum()
(county_pct_change.isin([np.inf]).sum() == 1).sum()
(county_pct_change.isin([np.inf]).sum() == 2).sum()
(county_pct_change.isin([np.inf]).sum() == 3).sum()
```

```python
# Limited number of infinity values per column
# Happens when jump from 0 to x number of cases,
# Could happen multiple times if num cases drops and then 
# increases or rate changes

# Replace infinities with -100 --> they are the start
county_pct_change = county_pct_change.replace([np.inf], -100)
county_pct_change.tail()
```

```python
# Now get the max value for each
county_pct_change.max(axis = 0).isna().sum()
# 388 counties have no change
max_pct_change = county_pct_change.max(axis = 0)
max_pct_change
```

```python
county_pct_change.idxmax()
```

```python
# Add max_pct_change to df and date when max occurred
max_df = pd.DataFrame([county_pct_change.max(axis = 0), 
                       county_pct_change.idxmax()])
max_df = max_df.transpose()
max_df = max_df.rename(columns={0: 'max_pct_change', 1:'max_date'})
max_df['State'] = max_df.index.get_level_values('State')
max_df['CountyName'] = max_df.index.get_level_values('CountyName')
max_df = max_df.reset_index(drop = True)
max_df
```

```python
# Merge county confirmed with max pct_change data
county_confirmed_pct_change = county_data_confirmed.merge(max_df, 
                                                          how = 'inner',
                                                          left_on = ['State', 'CountyName'],
                                                          right_on = ['State', 'CountyName'])
county_confirmed_pct_change.head()
```

```python
# Get rid of counties with max pct change na values
county_clean = county_confirmed_pct_change.dropna(subset=['max_pct_change'])
county_clean.head()
```

### Split data into training and testing sets

```python
train_pct, test_pct = sklearn.model_selection.train_test_split(
    county_clean, 
    train_size = 0.8, test_size = 0.2, random_state = 19)
print('Train:', train_pct.shape, 'Test:', test_pct.shape)
```

### Data processing and modeling

```python
def select_columns(data, *columns):
    return data.loc[:, columns]

def process_data_county(data):
    data = select_columns(data, 
                          '4/1/20',
                          '4/2/20',
                          '4/3/20',
                          '4/4/20',
                          '4/5/20',
                          '4/6/20', 
                          '4/7/20',
                          '4/8/20', 
                          '4/9/20', 
                          '4/10/20',
                          '4/11/20',
                          '4/12/20',
                          '4/13/20',
                          '4/14/20',
                          '4/15/20',
                          '4/16/20',
                          '4/17/20',
                          '4/18/20',
                          'PopulationDensityperSqMile2010',
                          'max_pct_change',
                          '#HospParticipatinginNetwork2017',
                          '#Hospitals',
                          '#ICU_beds')
    
    X = data.drop(['4/18/20'], axis = 1)
    y = data.loc[:, '4/18/20']
    
    return X, y
```

```python
X_train, y_train = process_data_county(train_pct)
X_test, y_test = process_data_county(test_pct)

# X_train.head()
```

```python
from sklearn import linear_model as lm
linear_model = lm.LinearRegression(fit_intercept = True)
```

```python
linear_model.fit(X_train, y_train)
y_fitted = linear_model.predict(X_train)
y_fitted_test = linear_model.predict(X_test)
```

```python
def rmse(predicted, actual):
    """
    Calculates RMSE from actual and predicted values
    """
    return np.sqrt(np.mean((actual - predicted)**2))
```

```python
training_error = rmse(y_train, y_fitted)
test_error = rmse(y_test, y_fitted_test)
print("Training RMSE: {}".format(training_error))
print(test_error)
```

```python
from sklearn.model_selection import KFold
from sklearn.base import clone

def cross_validate_rmse(model, X, y):
    model = clone(model)
    five_fold = KFold(n_splits=5)
    rmse_values = []
    for tr_ind, va_ind in five_fold.split(X):
        model.fit(X.iloc[tr_ind,:], y.iloc[tr_ind])
        rmse_values.append(rmse(y.iloc[va_ind], model.predict(X.iloc[va_ind,:])))
    return np.mean(rmse_values)
```

```python
cv_error = cross_validate_rmse(linear_model, X_train, y_train) #is this right? idk
print("Cross Validation RMSE: {}".format(cv_error))
```

## Predicting total deaths by state

```python
def select_columns(data, *columns):
    return data.loc[:, columns]

def process_data(data):
    data = select_columns(data,
                          'Deaths',
                         'Confirmed',
                         'People_Tested')
    
    X = data.drop(['Deaths'], axis = 1)
    y = data.loc[:, 'Deaths']
    
    return X, y

def process_data_final(data):
    data = select_columns(data,
                          'Deaths',
                         'Confirmed',
                         'Incident_Rate',
                         'People_Tested',
                         'Testing_Rate')
    
    X = data.drop(['Deaths'], axis = 1)
    y = data.loc[:, 'Deaths']
    
    return X, y
```

```python
X_train, y_train = process_data(train_state)
X_test, y_test = process_data(test_state)
```

```python
from sklearn import linear_model as lm
linear_model = lm.LinearRegression(fit_intercept=True)

# Fit model
linear_model.fit(X_train, y_train)
# Compute fitted and predicted values of Deaths for a given state
y_fitted = linear_model.predict(X_train)
```

```python
def rmse(predicted, actual):
    """
    Calculates RMSE from actual and predicted values
    Input:
      predicted (1D array): vector of predicted/fitted values
      actual (1D array): vector of actual values
    Output:
      a float, the root-mean square error
    """
    return np.sqrt(np.mean((actual - predicted)**2))
```

```python
training_error = rmse(y_train, y_fitted)
print("Training RMSE: {}".format(training_error))
```

```python
from sklearn.model_selection import KFold
from sklearn.base import clone

def cross_validate_rmse(model, X, y):
    model = clone(model)
    five_fold = KFold(n_splits=5)
    rmse_values = []
    for tr_ind, va_ind in five_fold.split(X):
        model.fit(X.iloc[tr_ind,:], y.iloc[tr_ind])
        rmse_values.append(rmse(y.iloc[va_ind], model.predict(X.iloc[va_ind,:])))
    return np.mean(rmse_values)
```

```python
cv_error = cross_validate_rmse(linear_model, X_train, y_train)
print("Cross Validation RMSE: {}".format(cv_error))
```

```python
final_model = lm.LinearRegression(fit_intercept=True)
final_model.fit(X_train, y_train)
y_predicted_train = final_model.predict(X_train)
y_predicted_test = final_model.predict(X_test)
```

```python
training_rmse = rmse(y_predicted_train, y_train)
test_rmse = rmse(y_predicted_test, y_test)
print(training_rmse)
print(test_rmse)
```

```python
# Improved values
X_train, y_train = process_data_final(train_state)
X_test, y_test = process_data_final(test_state)

final_model.fit(X_train, y_train)
y_predicted_train = final_model.predict(X_train)
y_predicted_test = final_model.predict(X_test)

training_rmse = rmse(y_predicted_train, y_train)
test_rmse = rmse(y_predicted_test, y_test)
print(training_rmse)
print(test_rmse)

cv_error = cross_validate_rmse(linear_model, X_train, y_train)
print("Cross Validation RMSE: {}".format(cv_error))
```
